# Lcat Function

Alphabets (Î£) = {a,b}


Language (L1) = string of length 2

          L1 = {"aa","ab","ba","bb"}

Language (L2) = string of length 3

          L2 = {"aaa","aab","aba","baa","abb","bab","bba","bbb"}

Defining Concatenation Function

```python
def lcat(L1,L2):

     return {x+y for x in L1 for y in L2}
```

Result lcat function  :

```python
L1 = {"aa","ab","ba","bb"}
L2 = {"aaa","aab","aba","baa","abb","bab","bba","bbb"}
lcat(L1,L2)
```

## lstar Function

Define lstar function

```python
def lstar(L1,n):
    
    star=set()
    for k in range(n+1):
        if k == 0:
            star.add("")
        else:
            new =lcat(star,L1)
            star.update(new)
    return star
```

Result lstar function

```python
lstar(L1,2)
```
